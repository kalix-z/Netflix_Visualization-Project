```python
import pandas as pd

df=pd.read_csv('netflix_titles.csv')
print(df.head())
print(df.info())
print(df.isnull().sum())
```

```python
df.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Convert and prepare data
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year

sns.set(style='whitegrid')  # global style
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type', hue='type', palette='coolwarm', legend=False)
plt.title('Movies vs TV Shows on Netflix', fontsize=14)
plt.xlabel('')
plt.ylabel('Count')
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(8,4))
sns.barplot(
    x=top_countries.values,
    y=top_countries.index,
    hue=top_countries.index,
    palette='crest',
    legend=False
)
plt.title('Top 10 Countries with Most Titles', fontsize=14)
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(
    data=df,
    x='rating',
    hue='rating',
    order=df['rating'].value_counts().index[:10],
    palette='viridis',
    legend=False
)
plt.title('Top 10 Ratings on Netflix', fontsize=14)
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(
    data=df,
    x='year_added',
    hue='year_added',
    palette='mako',
    order=sorted(df['year_added'].dropna().unique()),
    legend=False
)
plt.title('Netflix Content Added per Year', fontsize=14)
plt.xlabel('Year Added')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
genres = df['listed_in'].value_counts().head(10)

plt.figure(figsize=(10,4))
sns.barplot(
    x=genres.values,
    y=genres.index,
    hue=genres.index,
    palette='cubehelix',
    legend=False
)
plt.title('Top 10 Most Common Genres on Netflix', fontsize=14)
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.show()
```
